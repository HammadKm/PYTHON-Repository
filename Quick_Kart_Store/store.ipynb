```python
class Product:
    def __init__(self, id, name, price):
        self.id = id
        self.name = name
        self.price = price
        
class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def list_products(self):
        return self.products
       

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, product, quantity):
        self.items.append((product, quantity))


    def calculate_total(self):
        total = 0
        for product, quantity in self.items:
            total += product.price * quantity
        return total


inventory = Inventory()
cart = ShoppingCart()

while True:
    print("Choose an action:")
    print("1. Shop for products")
    print("2. Sell a product")
    print("3. Show Cart")
    print("4. Checkout")
    print("5. Quit")

    choice = input("Enter the number of your choice: ")

    if choice == "1":
        print("Available Products:")
        for product in inventory.list_products():
            print(f"{product.id}. {product.name} - ${product.price}")
        product_id = int(input("Enter the product ID you want to buy: "))
        selected_product = next((p for p in inventory.products if p.id == product_id), None)
        if selected_product:
            quantity = int(input("Enter the quantity: "))
            cart.add_item(selected_product, quantity)
            print(f"----{selected_product.name} added to cartðŸ›’ ------- ")
            print("")
        else:
            print("----Product not found âš  -----")
            print("")
            
    elif choice == "2":
        product_name = input("Enter the product name you want to sell: ")
        product_price = float(input("Enter the price: "))
        product = Product(len(inventory.products) + 1, product_name, product_price)
        inventory.add_product(product)
        print(f"----{product.name} added to the inventory------")
        print("")
        
    elif choice == "3":
        print("Cart:")
        for product, quantity in cart.items:
            print(f"{product.name} - ${product.price} x {quantity}")
        print(f"----ðŸ’µTotal: ${cart.calculate_total()}----")
        print("")
        
    elif choice == "4":
        print(f"---Total: ${cart.calculate_total()}----")
        print("Thank you for shopping with us!")
        print("")
        break
        
    elif choice == "5":
        print("GoodbyeðŸ‘‹!")
        print("")
        break
    else:
        print("Invalid option.ðŸ˜±")
        print("")
```
